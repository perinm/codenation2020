```python
import spacy
import pandas as pd
import time
import os
from spacy.lemmatizer import Lemmatizer

pd.options.display.max_columns = 500
pd.options.display.max_rows = 500
```

```python
def do_lemma(t):
    phrase = []
    # disable=['ner','textcat', 'tagger'] desabilitará
    # funcionalidades que não serão utilizadas no processo de 
    # lemmatização, deixando 'nlp' menos carregado e mais rápido.
    docs = nlp(str(t), disable=['ner','textcat', 'tagger'])
    for sent in docs.sents:
        phrase.append(sent.text)
        
    return " ".join(phrase)
```

```python
start_time = time.time()
l_doc=[]
for i in range(30):
    do_lemma(df["0"][:5])
    l_doc.append(nlp(df["0"][i]))
print("--- %s seconds ---" % (time.time() - start_time))
```

```python
start_time = time.time()
tokens1 = nlp(df["0"][0])
df_vazio=pd.DataFrame(columns=["id","similarity"])
for i in range(1000):
    df_vazio = df_vazio.append({'id':df['id'][i],'similarity':tokens1.similarity(l_doc[i])},ignore_index=True)
    print(i)
print("--- %s seconds ---" % (time.time() - start_time))
```
