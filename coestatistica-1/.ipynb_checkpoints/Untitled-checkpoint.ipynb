```python
import pandas as pd
import json
```

```python
df = pd.read_csv("desafio1.csv")
```

```python
df
```

```python
grupo = df.groupby("estado_residencia").pontuacao_credito.agg(["mean","median","std"])
```

```python
grupo = df.groupby("estado_residencia").agg(mediana=("pontuacao_credito","median"),
                                            media=("pontuacao_credito","mean"),
                                            desvio_padrao=("pontuacao_credito","std")
                                           )
```

```python
grupo
```

```python
dici = grupo.to_dict('index')
```

```python
dici
```

```python
for i in df.estado_residencia.unique():
    dici[i]["moda"]=int(df[df["estado_residencia"]==i].pontuacao_credito.mode()[0])
```

```python
dici
```

```python
with open('submission.json', 'w') as fp:
    json.dump(dici, fp,sort_keys=True, indent=4)
```
